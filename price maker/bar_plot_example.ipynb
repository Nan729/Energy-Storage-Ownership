```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
profit_df_summer = pd.DataFrame(columns=["species", "type","profit"])
value_df_summer = pd.DataFrame(columns=["species", "type","value"])
profit_df_winter = pd.DataFrame(columns=["species", "type","profit"])
value_df_winter = pd.DataFrame(columns=["species", "type","value"])
```

```python
#### data
# Load the arrays from the npz file
data = np.load('./Interresult/heatmap_solar_summer.npz')
m_joint = data ['m_joint'][-1,-1]*1000
m_disjoint = data ['m_disjoint'][-1,-1]*1000
m_single = data ['m_single'][-1,-1]*1000
m_joint_noArb = data ['m_joint_noArb'][-1,-1]*1000
valueWind = data ['valueWind'][-1,-1]*1000
valueWind_noArb = data ['valueWind_noArb'][-1,-1]*1000
value_arbitrager = data ['value_arbitrager'][-1,-1]*1000
print("solar_summer percentage", valueWind/m_single*100, valueWind_noArb/m_single*100, value_arbitrager/m_single*100)
profit_df_summer.loc[0] = ["Summer\nSolar", "Ori", m_single]
profit_df_summer.loc[1] = ["Summer\nSolar", "Disjoint", m_disjoint]
profit_df_summer.loc[2] = ["Summer\nSolar", "Joint", m_joint]
profit_df_summer.loc[3] = ["Summer\nSolar", "Joint (mono)", m_joint_noArb]
value_df_summer.loc[0] = ["Summer\nSolar", "RES", valueWind]
value_df_summer.loc[1] = ["Summer\nSolar", "RES (mono)", valueWind_noArb]
value_df_summer.loc[2] = ["Summer\nSolar", "Arbitrageur", value_arbitrager]

data = np.load('./Interresult/heatmap_wind_summer.npz')
m_joint = data ['m_joint'][-1,-1]*1000
m_disjoint = data ['m_disjoint'][-1,-1]*1000
m_single = data ['m_single'][-1,-1]*1000
m_joint_noArb = data ['m_joint_noArb'][-1,-1]*1000
valueWind = data ['valueWind'][-1,-1]*1000
valueWind_noArb = data ['valueWind_noArb'][-1,-1]*1000
value_arbitrager = data ['value_arbitrager'][-1,-1]*1000
print("wind_summer percentage", valueWind/m_single*100, valueWind_noArb/m_single*100, value_arbitrager/m_single*100)
profit_df_summer.loc[4] = ["Summer\nWind", "Ori", m_single]
profit_df_summer.loc[5] = ["Summer\nWind", "Disjoint", m_disjoint]
profit_df_summer.loc[6] = ["Summer\nWind", "Joint", m_joint]
profit_df_summer.loc[7] = ["Summer\nWind", "Joint (mono)", m_joint_noArb]
value_df_summer.loc[3] = ["Summer\nWind", "RES", valueWind]
value_df_summer.loc[4] = ["Summer\nWind", "RES (mono)", valueWind_noArb]
value_df_summer.loc[5] = ["Summer\nWind", "Arbitrageur", value_arbitrager]
# fig = plt.figure(figsize=(4, 3))
g = sns.catplot(
    data=value_df_summer, kind="bar",
    x="species", y="value", hue="type",
    errorbar=None, palette="dark", alpha=.6
)
g.fig.set_size_inches(6, 3)
# g.despine(left=True)
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):.0e}'))
g.set_axis_labels("", "Profit ($)")
g.legend.set_title("")
plt.show()
```

```python
#### data
# Load the arrays from the npz file
data = np.load('./Interresult/heatmap_solar_winter.npz')
m_joint = data ['m_joint'][-1,-1]*1000
m_disjoint = data ['m_disjoint'][-1,-1]*1000
m_single = data ['m_single'][-1,-1]*1000
m_joint_noArb = data ['m_joint_noArb'][-1,-1]*1000
valueWind = data ['valueWind'][-1,-1]*1000
valueWind_noArb = data ['valueWind_noArb'][-1,-1]*1000
value_arbitrager = data ['value_arbitrager'][-1,-1]*1000
print("solar_winter percentage", valueWind/m_single*100, valueWind_noArb/m_single*100, value_arbitrager/m_single*100)
profit_df_winter.loc[0] = ["Winter\nSolar", "Ori", m_single]
profit_df_winter.loc[1] = ["Winter\nSolar", "Disjoint", m_disjoint]
profit_df_winter.loc[2] = ["Winter\nSolar", "Joint", m_joint]
profit_df_winter.loc[3] = ["Winter\nSolar", "Joint (mono)", m_joint_noArb]
value_df_winter.loc[0] = ["Winter\nSolar", "RES", valueWind]
value_df_winter.loc[1] = ["Winter\nSolar", "RES (mono)", valueWind_noArb]
value_df_winter.loc[2] = ["Winter\nSolar", "Arbitrageur", value_arbitrager]


data = np.load('./Interresult/heatmap_wind_winter.npz')
m_joint = data ['m_joint'][-1,-1]*1000
m_disjoint = data ['m_disjoint'][-1,-1]*1000
m_single = data ['m_single'][-1,-1]*1000
m_joint_noArb = data ['m_joint_noArb'][-1,-1]*1000
valueWind = data ['valueWind'][-1,-1]*1000
valueWind_noArb = data ['valueWind_noArb'][-1,-1]*1000
value_arbitrager = data ['value_arbitrager'][-1,-1]*1000
print("wind_winter percentage", valueWind/m_single*100, valueWind_noArb/m_single*100, value_arbitrager/m_single*100)
profit_df_winter.loc[4] = ["Winter\nWind", "Ori", m_single]
profit_df_winter.loc[5] = ["Winter\nWind", "Disjoint", m_disjoint]
profit_df_winter.loc[6] = ["Winter\nWind", "Joint", m_joint]
profit_df_winter.loc[7] = ["Winter\nWind", "Joint (mono)", m_joint_noArb]
value_df_winter.loc[3] = ["Winter\nWind", "RES", valueWind]
value_df_winter.loc[4] = ["Winter\nWind", "RES (mono)", valueWind_noArb]
value_df_winter.loc[5] = ["Winter\nWind", "Arbitrageur", value_arbitrager]
fig = plt.figure(figsize=(4, 3))
g= sns.catplot(
    data=value_df_winter, kind="bar",
    x="species", y="value", hue="type",
    errorbar=None, palette="dark", alpha=.6
)
# g.despine(left=True)
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):.0e}'))
g.set_axis_labels("", "Profit ($)")
g.legend.set_title("")
```

```python
df_combined = pd.concat([value_df_summer, value_df_winter], ignore_index=True)

g= sns.catplot(
    data=df_combined, kind="bar",
    x="species", y="value", hue="type",
    errorbar=None, palette="muted", alpha=.6, legend=False
)
# g.despine(left=True)
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):.0e}'))
g.set_axis_labels("", "Value ($)",fontsize=14)
# g.legend.set_title("")
g.fig.set_size_inches(8, 3)
ax = g.facet_axis(0, 0)

hatches = ['', 'xx', '/']
print(len(hatches))
# for i, bar in enumerate(ax.containers):
#     bar.set_hatch(hatches[i % len(hatches)])

for i, p in enumerate(ax.patches):
    print(i, hatches[i % len(hatches)])
    ax.annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')
    # p.set_hatch(hatches[i % len(hatches)])



# Get the underlying matplotlib axes object
ax = g.ax
# Move the legend inside the figure
ax.legend(loc='upper left', bbox_to_anchor=(0.7, 1.15), ncol=1, fontsize = 12)
g.set_xticklabels(fontsize=12)
g.set_yticklabels(fontsize=12)
plt.show()
g.tight_layout()
g.savefig("./pics/compare_value_bar.pdf")
```
